### Problem Statement <i>Synthesis</i>

It is hard for the company to give loans to people with no credit history. We work in a consumer finance company which specializes in lending various types of loans to urban customers. We have to analyze patterns in the data. This will ensure that the applicants are capable of repaying the loan are not rejected.

When a loan application is recieved, there are two main risks associated with the bank's decision:
- If the applicant is likely to repay the loan, then not approving the loan results in a loss of business to the company
- If the applicant is not likely to repay the loan, i.e. he/she is likely to default, then approving the loan may lead to a financial loss for the company

The two scenarios that we are are concerned with are:
- The client with payment difficulties
- Paid on Time (named "All other cases")

The four decisions that can be taken by client/ company comprise of the below selection:
1. <b>Approved:</b> The company has approved the loan application
2. <b>Cancelled:</b> The client cancelled the loan application sometime during the approval. (Bad pricing or changed their mind)
3. <b>Refused:</b> The company has refused the loan application
4. <b>Unused Offer:</b> Loan has been cancelled by the client, but on different stage of progress

### Business Objectives
This case study aims to identify patterns which indicate if a client has difficulty paying their instalments which may be used for taking actions such as denying the loan, reducing the amount of loan, lending (to risky applicants) at a higher interest rate, etc. This will ensure that the consumers capable of repaying the loan are not rejected. Identification of such applicants using EDA is the aim of this case study.

```python
# Defining a function to make boxplots for a specified dataframe
def plotBoxPlot(df, colName):
    '''
    This function is used to set the stype of the box plot, name the graph and plot the box plot for the specified column
    
    Inputs:
    @df (dataframe) - The dataframe for which boxplots are to be plotted
    @colName (string) - The numeric column for which box plot is to be plotted
    
    Output:
    Titles box plot of specified colName
    '''
    sns.set(style="whitegrid")
    plt.figure(figsize=(20,5))
    plt.title(colName)
    sns.boxplot(x=df[colName])
    
    #fig = px.box(df, x=colName)#, y="total_bill")
    #fig.show()
```

### Binning the continous variables

We shall now perform binning on a few columns. One way to do it is by using the in-built pd.cut()/ pd.qcut() functionality by  Pandas.

#### Cumulative Population Column


**Note:**
- REGION_POPULATION_RELATIVE --> Column indicating Normalized population. Post analysis from the boxplot, we shall bin the column into sparse, normal and densely populated areas that the client belongs to

## Why correlation <u>cannot</u> be taken as a metric for binary predictors

We have now done correlation analysis on the entire table. We notice that for binary predictors, correlation is not an optimal method to find the most predicted variables. 

- We observe from the result that we have 1 correlation with EXT_SOURCE_2, besides which all of the variables show no correlations
- A correlation above 0.5 or below -0.5 would be considered significant. Here, we notice that there is <b>no</b> strong independent feature that is correlated with the target variable
- If we were performing a binary classification problem, we can tackle the issue by using the Variable Importance Feature in Random Forest or XGBoost

##### Pearson correlation is used for continuous data, Thus, when applied on binary data (TARGET), we obtain relationshops that are not precise or necessarily correct

```python
# Function to plot histogram for numerical, univariate analysis
def plotHistogram(df, colName):
    '''
    This function is used to set the style of the plot, name the graph and plot the distribution for the specified column
    
    Inputs:
    @df (dataframe) - The dataframe for which histograms are to be plotted
    @colName (string) - The numeric column for which histograms is to be plotted
    
    Output:
    Titles distribution plot of specified colName
    '''
    sns.set(style="whitegrid")
    plt.figure(figsize=(20,5)) 
    plt.title(colName)
    plt.ylabel('Density', fontsize=14)
    sns.distplot(df[colName], kde=True)#, fit=stats.gamma)
```

**Note:** Here the y-axis represents the frequency density for each of the bins such that the area under the curve is equal to 1.

```python
# Function to plot histogram for numerical, univariate analysis
def plotHistogram(df, colName):
    '''
    This function is used to set the style of the plot, name the graph and plot the distribution for the specified column
    
    Inputs:
    @df (dataframe) - The dataframe for which histograms are to be plotted
    @colName (string) - The numeric column for which histograms is to be plotted
    
    Output:
    Titles distribution plot of specified colName
    '''
    sns.set(style="whitegrid")
    plt.figure(figsize=(20,5)) 
    plt.title(colName)
    plt.ylabel('Density', fontsize=14)
    sns.distplot(df[colName], kde=True)#, fit=stats.gamma)
```

**Note:** Here the y-axis represents the frequency density for each of the bins such that the area under the curve is equal to 1.

# Inferences From Exploratory Data Analysis

**1.** Overwhelming majority of credit applicants with risk involved are young working clients that earn 1-2 lakhs per month and have experience of 1.5 to 2 years, risk analysis should target this age group
<br /><br /> **2.** Majority of the applicants request for Credit between 5-6 lakhs for goods prices worth approximately 5 lakhs. These applicants are likely to default
<br /><br />**3.** Most of the applicants that leads to credit default have loan annuity between Rs.25000-30000.
<br /><br /> **4.** Defaulters are applicants that completely higher secondary education, own a house and are married. We can conclude that experience of owning capital can still lead to loan defaults
<br /><br />**5.** Applicants that are unaccompanied during application process overwhelming shows the possibility of defaulting. Risk analysis should be more focused among this group
<br /><br /> **6.** Applicants that are likely to pay the loan amount earn comparatively the same amount as those who default but these applicants are likely to have more job experience 
<br /><br />**7.** Applicants that pay the loans on time apply for lesser credit loan between Rs.200000 - Rs 350000 compared to the defaulters
<br /><br /> **8.** Credit loan amount of applicants in other cases are more inline with the price of the goods than that of defaulters
<br /><br />**9.** Majority of applicants in non-defaulters group are educated, married and owns a house or an apartment
<br /><br /> **10.** Among the applicants having diffulcy paying the loan amounts, a significant portion of those on maternity leaves have high loan annuity amounts
<br /><br />**11.** Applicants that complete an academic degree and high annuity amount are more likely to default. This maybe due to preexisting loans that the applicant may have paid off in the past.
<br /><br /> **12.** Senior Citizens and applicants of old-age group have higher loan amount and are more likely to default
<br /><br />**13.** Married applicants generally have the highest load credit amount and are more likely to belong to non-defaulters group
<br /><br /> **14.** A large number of applicants of young age group have lower credit amount and show promising signs of being a non-defaulter. By targeting these sectors through marketing, higher revenue can be generated
<br /><br />**15.** Comparing the previous application and current application data, most of the applicants apply for higher amount of loan in the current period and thus more likely to default
<br /><br /> **16.** There is no significant jump in the salary of defaulters between the current and previous application although they submit for relatively higher amount of loan. Applicants that apply for more loan without salary jump is a red flag
<br /><br /> **17.** Applicants with one or more children are more likely to pay the loan than applicants with no children. Defaulters also tend to not report the number of family members
<br /><br />**18.** Comparing previous and current application, we can infer that the unemployed customers that recently procure a job are more likely to pay the loans on time
<br /><br /> **19.** Applicants that move from Rented apartments to their own house/apartments between applications are more likely to be a non-defaulter
<br /><br />**20.** There is no different in trend for defaulters that belong to senior or old age group between current and previous applications
