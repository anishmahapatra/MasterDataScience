## Data Loading - Data1, GSDP, Data2

```python
#Importing the required modules and packages

import matplotlib.pyplot as plt; plt.rcdefaults()
from matplotlib.ticker import StrMethodFormatter
%matplotlib inline
import seaborn as sns
import pandas as pd
import numpy as np
import glob, os

pd.set_option('display.max_columns', 500)

import warnings
warnings.filterwarnings("ignore")
```

```python
# Filling in the path variable to pick up the .csv files from the 'Data Folder'
# This is just the GSDP data
path = './Data/*.csv'

# Reading all the paths to the csv files and initializing an empty list
all_files = glob.glob(path)
allFilesList = []

# For loop to ingest all csv files, transpose the data and append the same onto a master datatset

for filename in all_files:
    # Reading each file into a pandas dataframe with 'latin' encoding
    df = pd.read_csv(filename, encoding = 'latin', index_col = None, header = 0)
    
    df = df[['S.No.', 'Item', '2014-15']]
#     df.set_index('S.No.')
    
    # Transposing the data using transpose() function
    df = df.transpose()
    
    # Removing the first row that contains the headers from df
    new_header = df.iloc[0]
    df = df[1:]
    
    # Making first row as column
    df.columns = new_header
    
    # Getting the name of the state to add it as a new column in the master dataset
    df.loc[:, 'State'] = filename.split('\\')[-1].split('.')[0]
    allFilesList.append(df)
    
# Concatenating the data into the master dataset called dfGSDP
dfStateGDP = pd.concat(allFilesList, axis = 0, ignore_index = True)
```

```python
# Sense check of master dataset of State's GDP
dfStateGDP.head()
```

```python
# Removing the first row that contains the headers from df
new_header = dfStateGDP.iloc[0]
dfStateGDP = dfStateGDP[1:]

# Making first row as column
dfStateGDP.columns = new_header

# Renaming the "Andhra Pradesh" column to "States" to maintain naming nomenclature
dfStateGDP.rename(columns={'AndhraPradesh':'State'}, inplace=True)

# Reindexing the data
dfStateGDP.reset_index(inplace = True, drop = True)
```

```python
# Moving the "States" column to the start of the dataframe
# Get a list of columns
cols = list(dfStateGDP)

# Move the column to head of list using index, pop and insert
cols.insert(0, cols.pop(cols.index('State')))

# use ix to reorder
dfStateGDP = dfStateGDP.ix[:, cols]
```

## Best fit line - States (Plots)

```python
# Defining a user-defined function to take in the dataframe and the name of the state
# The purpose of the function is to plot a scatter plot as well as a best-fit line for the states

def scatterAndBestFitLine(df, stateName):
    y = df[stateName]
    plt.scatter(x = 'duration', y = stateName, data = df)
    plt.plot(np.poly1d(np.polyfit(y, np.arange(4), 1))(np.unique(y)),np.unique(y),  '--r')
    plt.gca().set_title(stateName)
    plt.show()

# Defining the canvas size and of the plots
fig = plt.figure(figsize=(40, 40))
fig.add_subplot(9, 3, 1)
 
# list of States for which scatter plot as well as best fit line is to be made
listOfStates = ['andhra_pradesh', 'arunachal_pradesh', 'assam', 'bihar', 'chhattisgarh', 'goa', 'gujarat', 'haryana', 
                'jammu_&_kashmir', 'jharkhand', 'karnataka', 'kerala', 'madhya_pradesh', 'meghalaya', 'odisha', 'sikkim', 
                'tamil_nadu', 'telangana',  'uttar_pradesh']

# Initializing a for loop to iterate through the states that have all 4 values populated
for i in listOfStates:
    scatterAndBestFitLine(df, i)
```

```python
# Defining a function to plot scatter plot and best-fit line for States that have 3 values populated
def scatterAndBestFitLine2(df, stateName):
    y = df[stateName]
    plt.scatter(x = 'duration', y = stateName, data = df)
    plt.plot(np.poly1d(np.polyfit(y, np.arange(3), 1))(np.unique(y)),np.unique(y),  '--r')
    plt.gca().set_title(stateName)
    plt.show()

# Defining the canvas and the figure size of the plots to be plotted 
fig = plt.figure(figsize=(30, 30))

# Removing the rows that have the NaN values
df1 = df.iloc[0:-1,:]
fig.add_subplot(9, 3, 1)

# List of states that have 3 values populated in their respective columns
listOfStates2 = ['himachal_pradesh', 'maharashtra', 'manipur', 'mizoram', 'nagaland', 'punjab', 'rajasthan', 'tripura']

# Initializing a for loop to iterate through listOfStates2
for i in listOfStates2:
    scatterAndBestFitLine2(df1, i)
```

```python
# Setting the index as the State to get sorted stacked bar chart

# Making a function to get the sum() aggregation of dataset

def makeAggregatedSum(df, segmentationType):
    
    # Making the state column the index
    df.set_index('State', inplace=True)
    
    # Converting the rest of the data types from string to float
    df = df.iloc[:, 1:34].astype(str).astype(float)
    
    # Assigning the segmentation type of C1, C2, C3 and C4 to the concerned dataframe
    df['Segmentation'] = segmentationType
    
    # Grouping by the segmentation type
    df = df.groupby(['Segmentation']).sum()
    return df

# For the purposes of this analysis, we shall be considering the sub-sectors, 
# so, the primary, secondary and tertiary columns will be removed
redundant_columns = ['Primary', 'Secondary', 'Tertiary']
```
